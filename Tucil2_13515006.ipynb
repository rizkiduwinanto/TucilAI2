```python
from sklearn import datasets
iris = datasets.load_iris()
from sklearn.naive_bayes import MultinomialNB
naive_bayes = MultinomialNB()
prediction = naive_bayes.fit(iris.data, iris.target).predict(iris.data)
naive_bayes.predict_proba(iris.data)
```

```python
from sklearn.datasets import load_iris
from sklearn import tree
import graphviz
iris = load_iris()
classifier = tree.DecisionTreeClassifier()
classifier = classifier.fit(iris.data, iris.target)
dot_data = tree.export_graphviz(classifier, out_file=None, feature_names=iris.feature_names, class_names=iris.target_names, filled=True, rounded=True, special_characters=True)
graph = graphviz.Source(dot_data)
graph.render
```

```python
# loading library
from sklearn.neighbors import KNeighborsClassifier

# instantiate learning model (k = 3)
knn = KNeighborsClassifier(n_neighbors=3)

#train and predict iris
knntrain = knn.fit(iris.data, iris.target)

print(knntrain.predict(iris.data))
```

```python
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier(solver='lbfgs', alpha=1e-5,hidden_layer_sizes=(5, 2), random_state=1)
clftrain = clf.fit(iris.data, iris.target)

print(clftrain.predict(iris.data))
```
